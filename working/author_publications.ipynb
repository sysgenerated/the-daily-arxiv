```python
import requests

def search_arxiv_by_author(author_name):
    base_url = 'http://export.arxiv.org/api/query?search_query='
    query = f'au:"{author_name}"'
    url = base_url + query + '&metadata_prefix=arXiv'

    response = requests.get(url)
    if response.status_code == 200:
        return response.text
    else:
        print(f'Error: {response.status_code}')
        return None

author_name = "Tran Duy Anh Le"
response_text = search_arxiv_by_author(author_name)
print(response_text)  # This will print the XML response containing the search results with only article subjects
```
